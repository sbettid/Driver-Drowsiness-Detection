# Folder Preparation

In this notebook, we prepare the training data we have decided to use for this project (taken from http://mrl.cs.vsb.cz/eyedataset). 

**NOTE**: the dataset contained in the repository, have been already prepared executing this notebook on the original dataset, that can be downloaded from the link provided before. 

The two main operations, undertaken in this notebook are:



1.   Splitting the original training data in two folders, one per classes
2.   Splitting further this folders, dividing them in a training folder and a validation one (always containing the two classes)

Why did we perform a two-step process? Actually, at the beginning we did not forecast that we would have needed such a compex structure in terms of training data. 



## Part 1

This part has been performed one of our laptop itself, with the trainind data directly available on disk. The goal of this part is to read the training data, contained in the subfolder of mrlEyes_2018, copying then each image in the corresponding class image (0 for closed eye and 1 for open eye). The two class folders have been already created in the file system.

We start by importing the useful libraries

```python
import os
import subprocess
```

And we then copy each image in the corresponding folder, based on the file name which contains the eye status as the 5th indicator. 

```python
images = []
labels = []
path = "mrlEyes_2018_01/s00"
for i in range (1,38):
    if i < 10:
        curr_path = path + "0" + str(i)
    else:
        curr_path = path + str(i)
    print("Analysing folder: ", curr_path)
    for filename in os.listdir(curr_path):
        img_path = curr_path + "/" + filename
        lab = int(filename.split("_")[4])
        if lab == 0:
            os.system('cp ' + img_path + ' mrlEyes_2018_01/0/' + filename)
        else:
            os.system('cp ' + img_path + ' mrlEyes_2018_01/1/' + filename)
```
